```python
from transformers import AutoModelForCausalLM, AutoTokenizer
import torch

model_path = "/content/drive/MyDrive/intellihack5_models/qwen2.5-3b-research-qa-lora.gguf"

# Load base model
base_model_name = "Qwen/Qwen2.5-3B"  # Check the correct model name on Hugging Face

tokenizer = AutoTokenizer.from_pretrained(base_model_name)
model = AutoModelForCausalLM.from_pretrained(base_model_name, torch_dtype=torch.float16, device_map="auto")

# Load LoRA adapter
from peft import PeftModel

model = PeftModel.from_pretrained(model, model_path)

# Function to generate response
def generate_response(prompt):
    inputs = tokenizer(prompt, return_tensors="pt").to("cuda")
    output = model.generate(**inputs, max_new_tokens=512)
    return tokenizer.decode(output[0], skip_special_tokens=True)

# Test response
print(generate_response("Hello, how are you?"))
```
